#### Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Lead Origin Analysis

```python
lead_origin_count = data['Lead Origin'].value_counts()
print(lead_origin_count)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Lead Origin', data=data, order=lead_origin_count.index)
plt.title('Lead Origin Distribution')
plt.xticks(rotation=45)
plt.show()
```

##### Geographical Insights

```python
country_leads = data['Country'].value_counts()
print(country_leads)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Country', data=data, order=country_leads.index)
plt.title('Lead Distribution by Country')
plt.xticks(rotation=45)
plt.show()
```

##### Campaign Effectiveness

```python
campaign_conversion = data.groupby('Campaign')['Enrolment Status'].value_counts(normalize=True).unstack()
print(campaign_conversion)
```

```python
campaign_conversion.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Campaign Effectiveness (Conversion Rates)')
plt.ylabel('Proportion of Leads')
plt.show()
```

#####  Lead Stage Analysis

```python
lead_stage_count = data['Lead Stage'].value_counts()
print(lead_stage_count)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Lead Stage', data=data, order=lead_stage_count.index)
plt.title('Lead Distribution by Stage')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(data['Lead Stage'], bins=10, kde=True)
plt.title('Distribution of Leads by Lead Stage')
plt.xlabel('Lead Stage')
plt.ylabel('Frequency')
plt.show()
```

##### Correlation with Enrollment

```python
enrollment_vs_verification = pd.crosstab(data['Mobile Verification Status'], data['Enrolment Status'])
print(enrollment_vs_verification)
```

```python
enrollment_vs_verification.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Enrollment Status by Mobile Verification')
plt.show()
```

##### Pie Chart (for Proportional Data)

```python
plt.figure(figsize=(8, 8))
plt.pie(country_leads, labels=country_leads.index, autopct='%1.1f%%', startangle=140)
plt.title('Lead Distribution by Country')
plt.axis('equal')  
plt.show()
```

#####  Heatmap (for Matrix Data)

```python
heatmap_data = pd.crosstab(data['State'], data['Lead Stage'])

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt='d')
plt.show()
```

#####  Line Chart (for Trend Analysis)

```python
data['Instance Date'] = pd.to_datetime(data['Instance Date'])


leads_over_time = data.groupby(data['Instance Date'].dt.date).size()


plt.figure(figsize=(10, 6))
plt.plot(leads_over_time.index, leads_over_time.values, marker='o')
plt.title('Trend of Leads Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Leads')
plt.grid(True)
plt.show()
```

##### Scatter Plot (for Relationship Analysis)

```python
data['Conversion'] = data['Enrolment Status'].apply(lambda x: 1 if x == 'Enrolled' else 0)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lead Origin', y='Conversion', data=data)
plt.title('Lead Origin vs. Conversion')
plt.xticks(rotation=45)
plt.show()
```

#####  Box Plot (for Distribution Comparison)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='Lead Stage', data=data)
plt.title('Distribution of Lead Stages by Country')
plt.xticks(rotation=45)
plt.show()
```

##### Violin Plot (for Distribution and Density)

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Country', y='Lead Stage', data=data)
plt.title('Distribution of Lead Stages by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Selecting relevant numeric columns for the pair plot
df_pairplot = data[['Submitted Applications', 'Paid Applications', 'Lead Stage', 'Enrolment Status']]

# Pair plot
sns.pairplot(df_pairplot, hue='Enrolment Status', diag_kind='kde')
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```

```python
# Box plot for Submitted Applications by Lead Origin
plt.figure(figsize=(12, 6))
sns.boxplot(x='Lead Origin', y='Submitted Applications', data=data)
plt.title('Submitted Applications by Lead Origin')
plt.xticks(rotation=45)
plt.show()
```

```python
# Strip plot for Paid Applications by Lead Origin
plt.figure(figsize=(12, 6))
sns.stripplot(x='Lead Origin', y='Paid Applications', data=data, jitter=True)
plt.title('Paid Applications by Lead Origin')
plt.xticks(rotation=45)
plt.show()
```
